    print("After function call ",int)

</div><br>

    def function(int):
        int.append('B')print("Inside function call",int)

<pre style="color: green; font-size:17px; display:inline; font-weight:bold;">Create a function that can intake integer or decimal value and do following operations:
        a. Logarithmic function (log x)
        b. Exponential function (exp(x))
        c. Power function with base 2(2**x)
        d. Square root
</pre>

<h3 style="color:lightcoral">Solution: </h3>

```python
import math as mt

def solution(x):
    logValue = mt.log(x)
    expValue = mt.exp(x)
    powerValue = mt.pow(2, x)
    squareRoot = x ** 0.5
    
    allValue = f"Log Value is {logValue} \nExponential Value is {expValue} \nPower value is {powerValue} \nSquare Root value is {squareRoot}"
    return allValue 
x = 9
print(solution(x))
```

<span style="color:darkred; font-size:19px; font-weight:bold;">Que [9].  </span><span style="color: green; font-size:17px; font-weight:bold;">Create a function that takes a full name as an argument and returns first name and last name.</span>

<h3 style="color:lightcoral">Solution: </h3>  
</div>

```python
def firstLastName(fullname):
    listName = fullname.split(' ')
    if len(listName) == 2:
        name = f"First Name is: {listName[0]} \nLast Name is: {listName[1]}"
        return name
    else: 
        return "Missing something in Name"
    
fullname = input("Enter your full name: ")
print(firstLastName(fullname))
```
